# Сценарий 1: Взаимодействие с жителем (Личный кабинет / Мобильное приложение)

Данный сценарий ориентирован на конечного потребителя услуг — собственника жилья. Взаимодействие происходит в двух основных режимах: пассивный контроль и активное реагирование.

**Автоматизация коммерческого учета:**

 Действие пользователя: Житель открывает мобильное приложение для проверки начислений

# Сценарий 2: Взаимодействие с диспетчером УК
Данный сценарий является основным для эксплуатации инженерной инфраструктуры квартала.

**Мониторинг инфраструктуры:**

  Действие пользователя: Диспетчер контролирует общую ситуацию на мнемосхеме квартала.

Работа системы: Серверное ПО анализирует тренды поступающих данных (например, плавное падение давления в системе холодного водоснабжения в течение нескольких суток).

Результат: Система формирует предупреждение о высокой вероятности скрытой утечки на конкретном участке. Диспетчер направляет бригаду на диагностику до наступления аварии, что соответствует логике предиктивного обслуживания.



**Обработка критических инцидентов:**

Действие пользователя: Прием и верификация тревожного сигнала.


 Инцидент: Срабатывание датчика дыма в подъезде.

Работа системы: ПО перекрестно проверяет показания соседних датчиков (температуры, задымления) для исключения ложного срабатывания.

Результат: При подтверждении угрозы диспетчер видит на карте точный адрес и получает чек-лист действий: «Вызов МЧС», «Запуск голосового оповещения», «Разблокировка электромагнитных замков». *Курсив*

#### **4. Входные данные и их источники**


Какие типы данных будут использоваться в качестве входных данных для вашего решения? В каком формате они будут представлены и из каких источников планируется их получение?



Для работы системы используются следующие типы данных:

**1. Показания с датчиков (основной поток)**
- **Что это:** Потребление воды/света/тепла, температура батарей, давление в трубах, сигналы "протечка/дым/открыто".
- **Откуда берем:** С беспроводных датчиков, установленных в квартирах, подвалах, на чердаках.
- **В каком виде:** Датчики передают "сырые" цифровые пакеты, сервер их расшифровывает и превращает в понятные числа.



**2. Справочная информация (паспортные данные)**
- **Что это:** Какой датчик где висит (адрес, этаж, квартира), его заводской номер, когда проходил поверку.
- **Откуда берем:** Вручную забивает инженер при монтаже, плюс данные из проектной документации домов.
- **В каком виде:** Таблицы в базе данных (Excel-подобные списки).

**3. Внешние данные (для умных алгоритмов)**
- **Что это:** Погода на улице (температура, ветер), графики плановых отключений воды/света от горслужб.
- **Откуда берем:** С сайтов погоды (API), от ресурсоснабжающих организаций.
- **В каком виде:** Автоматические запросы через интернет в формате JSON (современный "почтовый конверт" для данных).

**4. Служебная информация**
- **Что это:** Уровень заряда батареек датчиков, качество сигнала (чтобы знать, не пора ли менять элемент питания или ставить усилитель).
- **Откуда берем:** Сам датчик присылает эти данные вместе с основными показаниями.
